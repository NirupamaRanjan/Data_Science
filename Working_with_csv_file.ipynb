<h1>2.Reading data from some URL

```python
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

<h1>3. sep parameter</h1>

<h1>4. Column Name</h1>

<h1>5.Index_col parameter</h1>

```python
df=pd.read_csv('/content/aug_train.csv',index_col='enrollee_id')
df
```

<h1>6. Header Parameter</h1>

</h3>

```python
df=pd.read_csv('/content/test.csv',header=1) #First row is treated like the header
df
```

<h1>7. usecols parameter</h1>

<h1>8.skiprows and nrows</h1>

<h1>9.Encoding parameter</h1>
<h3>Sometimes the encoding of the dataset is not UTF-8. So the first method is to chsnge its encoding in some code editor. Second method is to find the encoding and pass it to the encoding parameter.</h3>
<h1>10. Skip bad lines </h1>
<h3>In the dataset some rows might not satisfy the conditions of the dataset. These rows are removed by using a parameter error_bad_lines=False. There is also a parameter called on_bad_lines='skip' to skip bad rows</h3>

```python
df=pd.read_csv('/content/zomato.csv',usecols=['Restaurant ID','Restaurant Name','Address'],nrows=100,encoding='latin-1',on_bad_lines='skip')
df
```

<h1>11. dtypes parameter </h1>
<h3>We can change the datatype of a column</h3>

```python
df.info()
```

```python
df=pd.read_csv('/content/zomato.csv',usecols=['Restaurant ID','Restaurant Name','Address'],nrows=100,encoding='latin-1',on_bad_lines='skip',dtype={'Restaurant ID': float})
df.info()
```

<h1>12.Handling Dates</h1>
<h3>The time column read by pandas in csv file is in string format. To read them as date we use parse_dates parameter.Read documentation of this parameter for more usecases.</h3>

```python
pd.read_csv('/content/IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

<h1>13.Converters</h1>
<h3>If we need to change the value of rows based on certain conditions we use converters</h3>

```python
def rename(name):
  if name=='Chennai Super Kings':
    return 'CSK'
  return name

pd.read_csv('/content/IPL Matches 2008-2020.csv',parse_dates=['date'],converters={'team2':rename})
```

<h1>14. na_values parameter</h1>
<h3>Values you want to convert to NaN, empty. Sometimes you get Na_n,n_an. so you want to change them to NaN </h3>

```python
pd.read_csv('/content/aug_train.csv',na_values=["Male"])
```

<h1>15. Loading a huge dataset in chunks</h1>

```python
dfs=pd.read_csv('/content/aug_train.csv',chunksize=5000)
```

```python
#now every operation on the dataset will be applied on chunks in a loop.
for chunk in dfs:
  print(chunk.shape)
```
