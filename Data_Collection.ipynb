<h1>1. Fetching Data From API</h>

```python
# API KEY : 67868d6d18b9d0d8968e3a70d0d04226

import numpy as np
import pandas as pd
import requests as re
```

```python
  df=pd.DataFrame()
  for i in range(500):
    url='https://api.themoviedb.org/3/movie/popular?api_key=67868d6d18b9d0d8968e3a70d0d04226&page='+str(i+1)
    response=re.get(url)
    response=response.json()
    data=response['results']
    newdf=pd.DataFrame(data,columns=['adult','original_language','overview','popularity','release_date','title','vote_average','vote_count'])
    df = pd.concat([df, newdf])

```
