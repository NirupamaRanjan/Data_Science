```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('seaborn')

data=pd.read_csv('Scatter_plot_data.csv')
view_count=data['view_count']
likes=data['likes']
ratio=data['ratio']


# s = size of the dots
# c = color of the dots
plt.scatter(view_count, likes, s=100, c=ratio,cmap='summer',
            edgecolor='black' ,linewidth=1 , alpha=0.75)

cbar=plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Popular videos on youtube')
plt.xlabel('View_count')
plt.ylabel('Total_likes')

plt.show()


```
