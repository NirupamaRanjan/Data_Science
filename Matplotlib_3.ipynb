```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
#pie chart is useful when we have small dataset like 4 or 5 fields. Large dataset is often visualized by a bar graph

#data for pie chart
data=[50,60,80,150]

#labels on pie chart
labels=["data1", "data2", "data3" ,"data4"]

#colors for the pie chart
colors=["#ADD8E6","#FFC0CB","#DBDFAA","#73A9AD"]

plt.pie(data, labels=labels, wedgeprops={'edgecolor': 'black'}, colors=colors)
plt.title("Pie chart")

plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#Javascript will be exploded out by 20% of the radius
explode=[0.2,0,0,0,0]

plt.pie(slices, wedgeprops={'edgecolor':'black'} ,labels=labels , explode=explode ,shadow=True ,startangle=90 ,autopct='%1.1f%%')
plt.title("Popular Programming Languages")
plt.show()
```
